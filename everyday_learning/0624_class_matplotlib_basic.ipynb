```python
# 여러 개의 그래프 그리기

fig = plt.figure(figsize=(20, 12))
axs = []
for i in range(1, 10):
    ax = fig.add_subplot(3,3,i)
    axs.append(ax)
plt.show()
```

```python
## plot, bar 그래프 그리기

fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

x = range(0, 100)
y = [v * v for v in x]

ax1.plot(x, y)
ax2.bar(x, y)
plt.show()
```

```python
x = np.arange(0.0, 2 * np.pi, 0.1)

sin_y = np.sin(x)
cos_y = np.cos(x)

fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')

plt.show()
```

```python
tan_y = np.tan(x)
tanh_y = np.tanh(x)

fig = plt.figure(figsize=(12 , 8))

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)


ax1.plot(x, sin_y)
ax2.plot(x, cos_y)
ax3.plot(x, tan_y)
ax4.plot(x, tanh_y)

plt.show()
```

```python
x = np.arange(10)

plt.plot(x, 1 * x, label="y = %ix" % 1)
plt.plot(x, 2 * x, label="y = %ix" % 2)
plt.legend()

plt.show()
```

```python
# 5개의 그래프 그려보기

x = np.arange(10)

for i in range(5):
    plt.plot(x, i * x, label="y = %ix" % i)
plt.legend()
plt.show()
```
