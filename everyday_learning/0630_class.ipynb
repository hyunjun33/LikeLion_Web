```python
import seaborn as sns

iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.iplot(kind="bar", x="species", y="sepal_length")
```

```python
# 데이터 실습 2-1

iris.iplot(kind="bar", x="species", y="sepal_width", xTitle="Species", yTitle="sepal_width")
```

```python
iris.iplot(kind="bar", x="species", y="petal_length", xTitle="species", yTitle="petal_length")
```

```python
average = iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]].mean()
columns_name = iris.columns[:4]

average.iplot(kind="bar")
```

```python
iris["mean"] = iris.iloc[:, 0:4].mean(axis=1)
iris
```

```python
iris.iplot(kind="bar", x="species", y="mean")
```
