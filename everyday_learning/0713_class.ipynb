```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X, y = mglearn.datasets.make_forge()

X.shape, y.shape
```

```python
X[:10], y[:10]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    # fit 메소드는 self 오브젝트를 리턴합니다
    # 객체 생성과 fit 메소드를 한 줄에 쓸 수 있습니다
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} 이웃".format(n_neighbors))
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
axes[0].legend(loc=3)
```

```python
train_acc = []
test_acc = []
n_numbers = range(1, 22, 2) # 1, 3, 5 ~ 21 홀수만 적용해보기

for n in n_numbers:
    model = KNeighborsClassifier(n_neighbors=n)
    model.fit(X_train, y_train)
    
    acc_train = model.score(X_train, y_train)
    acc_test = model.score(X_test, y_test)
    
    train_acc.append(acc_train)
    test_acc.append(acc_test)
    
    print("k : ", n)
    print("학습 셋 정확도 : {:.3f}".format(acc_train))
    print("테스트 셋 정확도 : {:.3f}".format(acc_test))
```

```python
result = {"train_acc": train_acc, "test_acc": test_acc}
result_df = pd.DataFrame(result, index=np.arange(1, 22, 2))
result_df
```

```python
sns.lineplot(data=result_df, palette="tab10")
plt.show()
```

```python
# tips 데이터 셋에 대하여 학습 및 테스트 셋 점수를 그래프로 시각화 해보기

tips = sns.load_dataset("tips")
tips
```

```python
X_train, X_test, y_train, y_test = train_test_split(tips[["total_bill", "tip"]], tips["size"], stratify=tips["size"],
                                                    random_state=0)

train_acc = []
test_acc = []
k_numbers = []

for i in range(1, 100, 2):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)
    
    acc_train = model.score(X_train, y_train)
    acc_test = model.score(X_test, y_test)
    
    train_acc.append(acc_train)
    test_acc.append(acc_test)
    k_numbers.append(i)
    
    print("k :", i)
    print("학습 셋의 정확도 : {:.4f}".format(acc_train))
    print("테스트 셋의 정확도 : {:.4f}".format(acc_test))
```

```python
result = {"train_acc": train_acc, "test_acc": test_acc}
result_df = pd.DataFrame(data=result, index=k_numbers)
sns.lineplot(data=result_df, palette="tab10")
plt.show()
```
