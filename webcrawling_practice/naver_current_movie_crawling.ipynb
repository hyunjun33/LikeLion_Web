```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = "https://movie.naver.com/movie/running/current.nhn"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
```

```python
# 제목 가져오기
titles = []

table = soup.find("div", class_="lst_wrap")
movies = table.find_all("dt", class_="tit")

for i in range(len(movies)):
    title = movies[i].find("a").string
    titles.append(title)

print(len(titles))
print(titles)
```

```python
# 영화 평점 가져오기

scores = []

scores_tag = table.find_all("div", class_="star_t1")
scores_tag
for score in scores_tag:
    try:
        number = score.find("a").find("span", class_="num").string
    except:
        continue
    scores.append(number)

print(len(scores))
print(scores)
```

```python
# 참여자수

number_of_people = []
table = soup.find("div", class_="lst_wrap")
scores_tag = table.find_all("div", class_="star_t1")


for info in scores_tag:
    try:        
        participant = info.find("span", class_="num2").text[2:-1].strip()
    except:
        continue
    number_of_people.append(participant)

print(len(number_of_people))
print(number_of_people)
```

```python
# 예매율 가져오기

book_percent = []
booking_div = table.find_all("div", class_="star_t1 b_star")

for booking in booking_div:
    try:
        outcome = booking.find("span", class_="num").string + "%"
    except:
        continue
    book_percent.append(outcome)

print(len(book_percent))
```

```python
# 감독 이름

directors = []

movies = table.find_all("dl", class_="info_txt1")

for movie in movies :
    director = movie.find_all("dd")[1].find("a").string
    directors.append(director)
print(directors)
```

```python
# 크롤링한 정보 취합해서 딕셔너리로 취합 키는 영화이름, 값은 [이름, 평점, 참여자수, 감독, 예매율]

current_movies = {}

for i in range(len(titles)):
    information = []
    information.append(titles[i])
    information.append(scores[i])
    information.append(number_of_people[i])
    information.append(directors[i])
    try:
        information.append(book_percent[i])
    except:
        information.append('0%')
    current_movies[titles[i]] = information

print(len(current_movies))
print(current_movies)
```

```python
# 딕셔너리 데이터 문서화

df = pd.DataFrame(current_movies)
df = df.transpose()

df_final = df.rename(columns={0:"제목", 1:"평점", 2:"평점 참여자수", 3:"감독", 4:"예매율"})
df_final = df_final.reset_index(drop=True)
df_final.to_excel("네이버 현재 상영 영화 정보.xlsx")
```
