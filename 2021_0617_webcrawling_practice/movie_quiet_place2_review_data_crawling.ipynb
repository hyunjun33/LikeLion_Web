```python
import time
import sys
import io
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
from wordcloud import WordCloud
from PIL import Image
```

```python

base_url = "https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=189368&target=after&page=1"

comments = []
number_of_review = 0
for i in range(1, 33):
    page = urlopen(base_url + str(i))
    soup = BeautifulSoup(page, "html.parser")

    table_info = soup.find("table", class_="list_netizen")
    table_list = table_info.find_all("td", class_="title")
    for review in table_list:
        comment = review.find("a", class_="report").attrs["onclick"].split(",")[2].strip()
        comments.append(comment)
        number_of_review += 1
        
    print(comments)
    time.sleep(3)
```

```python
print("가져온 리뷰의 총 개수는 {}개 입니다.".format(number_of_review))
```

```python
dict_document = {"review": comments}
doc = pd.DataFrame(dict_document)
doc.head()
```

```python
doc.to_csv("콰이어트 플레이스 리뷰.csv", index=False)
f = open("콰이어트 플레이스 리뷰.csv", encoding="utf-8")
review = f.read()
f.close()
```
