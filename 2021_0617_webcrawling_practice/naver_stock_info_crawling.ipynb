```python
content_tag = soup.find("div", id="content")
content_tag
```

```python
# content 태그의 자녀 태그들 확인해보기.

flag = 1
for tag in content_tag.findChildren():
    print(tag)
    print('-------------------------------------------------', flag)
    flag += 1
```

```python
# 특정 종목 주식의 오늘 시세 값 가져오기

children_info = content_tag.findChildren()
today_price = children_info[9].find("span", class_="blind").text
print("오늘 시세: {}".format(today_price))
```

```python
count = 0
for tag in soup.find("p", class_="no_today").children:
    print(tag)
    print("----- 태그 확인 -----")
    count += 1
print("태그 개수: {}".format(count)) # \n, <em></em>, \n 줄바꿈도 태그로 인식한다?
```

```python
# 5개 주식 종목의 가격과 거래량 가져오기

stock_name = ['삼성전자', '카카오', 'NAVER', '카카오게임즈', '오리온']
stock_code = ['005930', '035720', '035420', '293490', '271560']
stock_price = []
stock_transaction_amount = []
stock_info = {}

base_url = "https://finance.naver.com/item/main.nhn?code="

for i, code in enumerate(stock_code):
    
    all_url = base_url + code
    page = urlopen(all_url)
    soup = BeautifulSoup(page, "lxml")
    
    tmp = soup.find("p", class_="no_today")
    today_price = tmp.find("span", class_="blind").text
    
    table_info = soup.find("table", class_="no_info")
    transaction_amount = table_info.find_all("td")[2].find("span", class_="blind").text
    
    stock_price.append(today_price)
    stock_transaction_amount.append(transaction_amount)

print(stock_name)
print(stock_price)
print(stock_transaction_amount)
```

### 국내 주식 종목 엑셀 파일을 이용해서 주식 종목의 값과 거래량을 크롤링하기

```python
import pandas as pd

stock_list = pd.read_excel("stock_list.xlsx")
stock_list.head()
```

```python
stock_names = stock_list['회사명']
stock_code = stock_list['종목코드']
```

```python
stock_names = stock_list['회사명']
stock_code = stock_list['종목코드']

stock_price = []
stock_transaction_amount = []
stock_info = {}

base_url = "https://finance.naver.com/item/main.nhn?code="

# 주식 종목 50개의 현재가와 거래량 가져오기

for i in range(50):
    
    all_url = base_url + str(stock_code[i])
    page = urlopen(all_url)
    soup = BeautifulSoup(page, "lxml")
    
    tmp = soup.find("p", class_="no_today")
    if tmp == None:
        continue
    today_price = tmp.find("span", class_="blind").text
   
        
    table_info = soup.find("table", class_="no_info")
    transaction_amount = table_info.find_all("td")[2].find("span", class_="blind").text
    
    price_transaction = []
    price_transaction.append(today_price)
    price_transaction.append(transaction_amount)
    stock_info[stock_names[i]] = price_transaction


print(stock_info)
```
