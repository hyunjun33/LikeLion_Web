```python
##인구수 가져오기

all_data = [] #최종 리스트
for i in range(2,10):
    tmp= '//*[@id="country-table"]/div/div/table/tbody/tr/td[{}]'.format(i)
    #print(tmp)
    sel_ele = driver.find_elements_by_xpath(tmp)
    
    column_data = []
    for one in sel_ele:
        #print(one.text)
        column_data.append(one.text)
        
    all_data.append(column_data)

#all_data
print(all_data)
```

```python
import pandas as pd
dict_dat = { "국가":all_data[0],
            "확진자":all_data[1],
            "사망자":all_data[2],
            "완치":all_data[3],
            "치명(%)":all_data[4],
            "완치(%)":all_data[5],
            "발생률":all_data[6],
            "인구수": all_data[7]}
data_df = pd.DataFrame(dict_dat)
data_df
```

```python
data_df.to_excel("covid_df.xlsx", index=True)
```
