```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import sys
import io
```

```python
url = "https://finance.naver.com/sise/"
page = urlopen(url)

soup = BeautifulSoup(page, "html.parser")

top_table = soup.find("table", id="siselist_tab_0")
top_table
```

```python
# 컬럼 인덱스 추출

columns_info = top_table.find("tr")
columns_names = columns_info.find_all("th")
column_index = []
for column in columns_names:
    column_index.append(column.text)
column_index.pop(0)
```

```python
# 1위부터 5위까지 정보

tr_list = top_table.find_all("tr")
stock_info = []
for i in range(2, 7):
    one_stock_info = []
    td_list = tr_list[i].find_all("td")
    for td in td_list:
        one_stock_info.append(td.text)
    stock_info.append(one_stock_info)
print(stock_info)
```

```python
# 6위 & 7위 정보

tr_list = top_table.find_all("tr")

for i in [10, 11]:
    one_stock_info = []
    td_list = tr_list[i].find_all("td")
    for td in td_list:
        one_stock_info.append(td.text)
    stock_info.append(one_stock_info)
print(stock_info)
```

```python
stock_final = {}
n = 1
for stock in stock_info:
    
    stock.pop(0)
    stock[4] = stock[4].replace("\n", "")
    stock[4] = stock[4].replace("\t", "")
    
    stock[5] = stock[5].replace("\n", "")
    stock[5] = stock[5].replace("\t", "")
    
    stock_final[n] = stock
    n += 1
    
print(stock_final)
```

```python
top_stock_df = pd.DataFrame(stock_final).transpose()
```

```python
top_stock_df.columns = column_index
top_stock_df
```

```python
top_stock_df.to_csv("Top 종목 상한가 정보.csv")
top_stock_df.to_excel("Top 종목 상한가 정보.xlsx")
```
